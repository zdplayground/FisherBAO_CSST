```python
sigma_photoz_list = [0.025, 0.05]
line_list = ['--', ':']
marker_list = ['x', 'd']

    
recon_dir = "pre_recon"
tracer = "photoz"
params_str = '11101010101010'
```

```python
alpha_perp_list = []
alpha_para_list = []

alpha_dir = idir0 + f"output/sigma_alpha/vary_params_{const_low}_{const_up}/Sigma_fog_{Sigma_fog}/{recon_dir}/k_width{kwidth}/Pspecz_sys{Pspecz_sys}/params_bf1.0/"
for sigma_photoz in sigma_photoz_list:
    filename = "sigma_alpha_{0}_tracer_zerror_specz{1:.3f}_photoz{2:.3f}_kmax{3:.2f}_params{4}.npz".format(tracer, sigma_specz, sigma_photoz, kmax, params_str)

    alpha_dir_temp = alpha_dir
    ifile = Path(alpha_dir, filename)
    data = np.load(ifile)
    alphas_mar = data['alphas_mar']

    alpha_perp_list.append(alphas_mar[:, 2])
    alpha_para_list.append(alphas_mar[:, 3])
```

## for $\sigma (\alpha_{\perp})$ and $\sigma (\alpha_{\parallel})$

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6), dpi=150)
plt.subplots_adjust(
    left=0.15,
    bottom=None,
    right=None,
    top=None,
    wspace=0.15,
    hspace=None,
)

i = 0
for sigalperp, sigalpara in zip(alpha_perp_list, alpha_para_list):
    axes[0].plot(z_mid, 100*sigalperp, ls=line_list[i], marker=marker_list[i], ms=8.0, color=f'C{i}')
    axes[1].plot(z_mid, 100*sigalpara, ls=line_list[i], marker=marker_list[i], ms=8.0, color=f'C{i}')
    i += 1

for j in range(2):
    axes[j].grid()
    axes[j].set_xlabel(r"$z_s$", fontsize=20)
    axes[j].set_yscale("log")
    
#axes[0].set_ylabel(r"$\sigma_{D_\mathrm{A}/r_{\mathrm{s}}}$ (%)", fontsize=20)
axes[0].set_ylabel(r"$\sigma (D_A(z)/r_d)$ (%)", fontsize=20)
axes[0].set_ylim([0.4, 30.0])
axes[0].set_yticks([0.5, 1.0, 5.0, 10.0])
axes[0].set_yticklabels(['0.5', '1.0', '5.0', '10.0'])

axes[1].set_ylabel(r"$\sigma (H(z)r_d)$ (%)", fontsize=20)

for i in range(2):
    axes[i].tick_params(axis='both', which='major', length=6, labelsize=14)
    axes[i].tick_params(axis='both', which='minor', length=3, labelsize=12)
        
#plt.tight_layout()
```

## also show the ratio between the two

```python
fig, axes = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]}, figsize=(8, 6), sharex=True, dpi=150)
plt.subplots_adjust(
    left=0.15,
    bottom=0.15,
    right=0.85,
    top=1.0,
    wspace=None,
    hspace=0.0,
)
for i, sigalperp in enumerate(alpha_perp_list):
    axes[0].plot(z_mid, 100*sigalperp, ls=line_list[i], marker=marker_list[i], ms=8.0, color=f'C{i}', label=r'$\sigma_z=%.3f(1+z)$'%(sigma_photoz_list[i]))
    axes[1].plot(z_mid, sigalperp/alpha_perp_list[0], ls=line_list[i], marker=marker_list[i], ms=6.0, color=f'C{i}')
    axes[i].grid(True)

    

axes[0].set_yscale("log")
axes[0].legend(fontsize=16)
axes[0].set_ylabel(r"$\sigma (D_A(z)/r_d)$ (%)", fontsize=20)
axes[0].set_ylim([0.5, 10.0])
axes[0].set_yticks([0.5, 1.0, 5.0, 10.0])
axes[0].set_yticklabels(['0.5', '1.0', '5.0', '10.0'])
axes[1].set_xlabel(r"$z_s$", fontsize=20)
axes[1].set_ylabel(r"$\sigma/\sigma_{\mathrm{fid}}$", fontsize=20)


for i in range(2):
    axes[i].tick_params(axis='both', which='major', length=6, labelsize=12)
    axes[i].tick_params(axis='both', which='minor', length=3, labelsize=10)
plt.tight_layout()

filename = f"photoz_sigma_DA_diff_photozerr_{zmin:.1f}z{zmax:.1f}.png"
ofile = odir + filename
plt.savefig(ofile)
```
