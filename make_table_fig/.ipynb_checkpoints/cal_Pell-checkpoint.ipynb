## Calculate $P_{\ell}$ from $P(k,\mu)$. --03-21-2023
\begin{align}
P_{\ell} = (2\ell+1)\int_0^1 P(k, \mu)L_{\ell}(\mu)d\mu
\end{align}

```python
%pylab inline
```

```python
from scipy import integrate
```

```python
# Pkmu with dimension (k, mu), mu in range [0, 1]
def cal_P0(Pkmu, mu, ell=0):
    return integrate.trapz(Pkmu, mu, axis=1)

def cal_P2(Pkmu, mu, ell=2):
    integrand = Pkmu * (3*mu**2.0-1)/2.0
    res = (2*ell+1)*integrate.trapz(integrand, mu, axis=1) 
    return res
```
